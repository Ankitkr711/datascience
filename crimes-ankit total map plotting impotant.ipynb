```python
df_murder.rename(columns={'Area_Name':'State','YEAR':'Year',
                       'Sub_Group_Name':'Gender',
                       'Upto 10 years':'Age < 10',
                       '10-15 years':'10 < Age < 15',
                       '15-18 years':'15 < Age < 18',
                       '18-30 years':'18 < Age < 30',
                       '30-50 years':'30 < Age < 50',
                       'Above 50 years':'Age > 50'
                       },inplace=True)
import warnings

warnings.filterwarnings('ignore')
```

```python
def assignVerticalAxis(df,g,value=1000):
     for i in range(0,g.axes.size):
        ax = g.fig.get_axes()[i]
        displayVerticalCount(df,ax,value)
def displayVerticalCount(df,ax,value):
    y_max = df.max() + value 
    ax.set_ylim(top=y_max)
for p in ax.patches:     
        if(math.isnan(p.get_height())):
            ax.text(p.get_x() + p.get_width()/2, 0, 0, 
                fontsize=12, color='Black', ha='center', va='bottom',rotation=90)
            continue
        else:            
            ax.text(p.get_x() + p.get_width()/2, p.get_height(), int(p.get_height()), 
                fontsize=12, color='Black', ha='center', va='bottom',rotation=90)
```

```python
for i in range(len(df_murder['Year'].value_counts().index)):
    temp = df_murder.drop(df_murder[df_murder['Gender'] == 'Male'].index)
    temp = temp.drop(temp[temp['Gender'] == 'Female'].index)
    temp = temp.drop(temp[temp['Year'] != temp['Year'].value_counts().index[i]].index)
    temp = temp.sort_values(by=['Victims_Total'], ascending=False)
    title = 'Murders in Different States for the year ' + str(temp['Year'].values[0])
    g = sns.factorplot(x='State',y='Victims_Total',data=temp,kind='bar',
                       palette=sns.color_palette('Reds_r',34),  
                       size=5,aspect=2).set(
                               title=title)
    assignVerticalAxis(temp['Victims_Total'],g)
    plt.xticks(rotation=90)
plt.show
```

```python
temp = df_murder.drop(df_murder[df_murder['State'] != 'Uttar Pradesh'].index)

sns.set(context='notebook',style='whitegrid',font_scale=1.3)

aa = temp.drop(temp[temp['Gender'] == 'Female'].index)
aa = aa.drop(aa[aa['Gender'] == 'Male'].index)

g = sns.factorplot(x='Year',y='Victims_Total',data=aa,               
               size=5,aspect=2).set(title='Murders in Uttar Pradesh over the years')
assignVerticalAxis(aa['Victims_Total'],g,500)
```

```python
temp = df_murder.drop(df_murder[df_murder['State'] != 'Uttar Pradesh'].index)
temp = temp.drop(temp[temp['Gender'] == 'Victims_Total'].index)

g = sns.factorplot(x='Year',y='Victims_Total',data=temp,kind='bar',hue='Gender',               
               size=5,aspect=2).set(title='Murders of Male and Female Population of "Uttar Pradesh" over the years')
assignVerticalAxis(temp['Victims_Total'],g)
```
